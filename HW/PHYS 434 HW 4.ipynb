```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

```python
t = np.linspace(0,np.pi * 100,int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
```

# Problem 1

```python
fig, ax = plt.subplots(1, 3, figsize=(30,10))

def custom_plot(start, num_points, step, plot_num):
    end = start + num_points
    ax[plot_num].plot(t[start:end:step], bkgd[start:end:step], 'o')

    
ax[0].set_title("Points 0-1000")
ax[1].set_title("Points 500-2500")
ax[2].set_title("Full Domain in Increments of 100")

custom_plot(0, 1000, 1, 0)
custom_plot(500, 2000, 1, 1)
custom_plot(0, len(t), 100, 2)
```

# Problem 2

```python
fig, ax = plt.subplots(1, 1)

g = ax.hist2d(t, bkgd, bins = 100, density = True)
ax.set_title("2D Histogram of Backgound Data")
ax.set_xlabel("Time")
ax.set_ylabel("Background Label")
fig.colorbar(g[3], ax = ax)
```

This plot shows the density of each value of the background at different t-values (the x-axis). The warmer a color (i.e. towards the top of the color bar), the more times values in that bin appear (each axis is divided into 100 bins of equal width).

# Problem 3

```python
new_t = t % (2*np.pi)
fig, ax = plt.subplots(1, 2, figsize = (24,12))

ax[0].plot(new_t, bkgd, 'o')

g = ax[1].hist2d(new_t, bkgd, bins = 100, density = True)
ax[1].set_title("2D Histogram of Backgound Data")
ax[1].set_xlabel("Folded Time")
ax[1].set_ylabel("Background Label")
fig.colorbar(g[3], ax = ax)
```
